```python
def extract_mcqs(link):
    data = requests.get(link)
    soup = BeautifulSoup(data.content, 'lxml')
    question = " ".join(soup.select('.question-body.mt-3.pt-3.border-top')[0].select('.card-text')[0].text.split())
    correct_options = []
    all_options = []

    no_of_all_options = len(soup.select('.question-body.mt-3.pt-3.border-top')[0].select('.multi-choice-item'))
    for i in range(no_of_all_options):
        all_options.append(" ".join(soup.select('.question-body.mt-3.pt-3.border-top')[0].select('.multi-choice-item')[i].text.split()))

    no_of_options = len(soup.select('.question-body.mt-3.pt-3.border-top')[0].select('.multi-choice-item.correct-hidden'))
    if no_of_options>1:
        for i in range(no_of_options):
            correct_options.append((" ".join(soup.select('.question-body.mt-3.pt-3.border-top')[0].select('.multi-choice-item.correct-hidden')[i].text.split()[0])))
    else:
        correct_options.append(" ".join(soup.select('.question-body.mt-3.pt-3.border-top')[0].select('.multi-choice-item.correct-hidden')[0].text.split()[0]))
    # random.shuffle(options) 
    return {
        "question": question,
        "options": all_options,
        "correct options": correct_options
    }

extract_mcqs("https://www.examtopics.com/discussions/amazon/view/145119-exam-aws-certified-data-engineer-associate-dea-c01-topic-1/")
```

```python
with open("DEA-C01 dumps.txt", 'r') as f:
    lines = f.readlines()
lines[1][3:]
```

```python
ques_ans = {}
num = 1
with open('questions answers.txt', 'w') as file:
    for line in lines[1:]:
        line = line[3:]
        response = extract_mcqs(str(line))

        ques_ans[response['question']] = response['options']
        ques_ans[response['question']].append(response['correct options'])
        question = response['question']

        file.write(str(num) + ". ")
        num+=1
        file.write(question)
        file.write('\n')
        # file.write(ques_ans[question])
        # print(ques_ans[question])
        correct_option = ques_ans[question].pop(-1)
        file.write("\t")
        file.writelines("\n\t".join(ques_ans[question]))
        file.write("\nCorrect option(s): ")
        file.writelines(correct_option)
        file.write("\n\n")

        # break

len(ques_ans)
```
